<b> Calculate incidence (7-days) rate </b><br>
<br>
In Germany: Incidence rate is a sum of daily (new) cases from the last 7 days per 100 000 inhabitants <br>
This rate is crucial for many decisions the government makes nowadays, e.g. whether a country is classified as a risk region or not

```python
# Divide the population through 100 000 to get results "per 100 000 inhabitants"

# Dataset without Kosovo
population_complete["div100"] = (population_complete["Population"] / 100000).round(2)
population_complete.head()
```

```python
# Dataset with Kosovo
population_complete_Kosovo["div100"] = (population_complete_Kosovo["Population"] / 100000).round(2)
population_complete_Kosovo.head()
```

```python
# Save population datasets to csv-files

population_complete.to_csv("C:/Users/Agnieszka/Downloads/Datasets/Covid_temp/population_complete.csv")
population_complete_Kosovo.to_csv("C:/Users/Agnieszka/Downloads/Datasets/Covid_temp/population_complete_Kosovo.csv")
```

```python
# Calculate 7-days incidence rate

# df = dataset with cases per day ("..._cpr2")

def incidence (df):
    
    # merge cpr dataset with population, only column "div100"
    df_global_population = pd.concat([df, population_complete_Kosovo["div100"]], axis=1, join="inner")
    
    # Calculate the sum of the last 7 days
    df_global_population["sum7days"] = df_global_population.iloc[:,-8:-1].sum(axis=1)
    
    # Calculate 7-days incidence
    df_global_population["incidence"] = (df_global_population["sum7days"] / df_global_population["div100"]).round(2)
    
    return df_global_population


#df["sum7days"] = df.iloc[:,-7:].sum(axis=1)
```

```python
confirmed_global_incidence = incidence(confirmed_global_cpr2)
confirmed_global_incidence.head()
```

```python
# Save the dataset with incidence rates to a csv file:

confirmed_global_incidence.to_csv("C:/Users/Agnieszka/Downloads/Datasets/Covid_temp/confirmed_global_incidence.csv")
```
