<b> Dataset preparation </b>

Overseas territories

```python
# Changing the name of Country/Region in case of overseas territories

def rename(df):
    
    overseas_terr=["Netherlands", "United Kingdom", "Denmark", "France"]
    for country in overseas_terr: 
        df["Country/Region"] = np.where((df["Country/Region"] == country) & (df["Province/State"].notna()), df["Province/State"], df["Country/Region"])
    return df
```

Function for summing up cases from all provinces for each country (Australia, China, Canada)

```python
# Function for summing up cases from all provinces 
# automatically for each country

# Selects countries with duplicates (countries which are devided into regions) and calculate a sum of cases

def sum_country(df):
    
    column_names = list(multiple_countries.index.values)
    sum_pro_country =[]
    for country in column_names:
        
        # Select rows with duplicates (countries which are devided into regions)
        mask = df["Country/Region"] == str(country)
        df_mask = df[mask]
        #return df_mask
    
        # Drop first few columns, leave only columns with number of cases
        df_data = df_mask.drop(columns=['Province/State', 'Country/Region', 'Lat', 'Long'], axis=0)
    
        # Reset index and create new one with number of rows
        ## Define index values(h)
        h = np.arange(len(df_data.values))
        df_data.reset_index()
        df_data2 = df_data.set_index(h)
        # return df_data2
 
        # Transpose dataframe to make summing easier
        df_data2_trans = df_data2.transpose()
    
        # Sum up data from each row and for each day, and store results in "m" (sum for each country stored as a list)
        m=[]
        c=0
        for c in range(len(df_data2_trans)): 
            x = df_data2.iloc[0:h[-1]+1, c].sum()
            c = c + 1
            m.append(x)
        #return m
        
        # Store results for each country in "sum_pro_country"        
        sum_pro_country.append(m)  
    
    # Convert results into a dataframe to use "to_list()" function
    sum_pro_country2 = pd.DataFrame([sum_pro_country]).transpose()
    
    # Split the lists in column "0" which include all data points as a list (comes from "m")
    # to get dataframes with country name as an index and dates as column names
    end_result = pd.DataFrame(sum_pro_country2[0].to_list(), columns=df_data2.columns, index=multiple_countries.index.values)
    
    return end_result
```

```python
# Prepare dataframes for further steps

def prepare (df):
    
    # remove columns and set "Country/Region" as an index
    df_short = df.drop(columns=['Province/State', 'Lat', 'Long'], axis=0).set_index("Country/Region")
    
    # Rename US to United States
    df_short.rename(index={'US':'United States'},inplace=True)
    
    # Sort index alphabetically
    df_short.sort_index(inplace=True)

    return df_short
```

Countries devided into provinces

```python
# Remove duplicates and replace values with a sum

# df = "..._short" dataframes, prepared dataset ("prepare" function)
# df_sum - "sum_..."- dataframes ("sum_country" function)

def remove_duplicates(df, df_sum):
    
    # Remove duplicates, keep only first value that will be replaced with the sum
    df_noIndex = df.reset_index()
    df_noDuplicates = df_noIndex.drop_duplicates(subset="Country/Region", keep="first")
    df_noDuplicates2 = df_noDuplicates.set_index("Country/Region")
    
    # Replace remained values with the sum of cases from each province ("sum_..."- dataframes)
    column_names = list(multiple_countries.index.values)
    for country in column_names:
        df_noDuplicates2.loc[country] = df_sum.loc[country]

    return df_noDuplicates2
```
